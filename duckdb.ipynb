```python
import duckdb
```

```python
with duckdb.connect("data.duckdb") as con:
    con.execute("""
        CREATE OR REPLACE TABLE film_permits AS
        SELECT *
        FROM read_csv_auto("Film_Permits_20260306.csv");
    """)
```

```python
with duckdb.connect("data.duckdb") as con:
    df = con.table("film_permits").to_df()
    print(df.head())
```
