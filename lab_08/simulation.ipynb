```python
latest_2025 = get_latest_data("data/PCPI25M2.csv").reset_index(drop=True)

append_df = show_table("cpi_append").reset_index(drop=True)
trunc_df = show_table("cpi_trunc").reset_index(drop=True)
inc_df = show_table("cpi_inc").reset_index(drop=True)

print("append matches latest:", append_df.equals(latest_2025))
print("trunc matches latest:", trunc_df.equals(latest_2025))
print("inc matches latest:", inc_df.equals(latest_2025))
print("row counts:", len(append_df), len(trunc_df), len(inc_df), len(latest_2025))
```

The results show that the three loading methods behave differently after revisions are introduced.

- `append` does not match the latest source file because it only adds new rows and does not overwrite revised historical values.
- `trunc` matches the latest source file because it reloads the full table each time.
- `incremental` also matches the latest source file because it updates changed rows and inserts new rows.

The row counts confirm this difference:
- `cpi_append`: 976 rows
- `cpi_trunc`: 937 rows
- `cpi_inc`: 937 rows
- latest source data: 937 rows

This shows that `append` kept outdated records, while `trunc` and `incremental` correctly reflected the latest CPI data.
